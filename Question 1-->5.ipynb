### COMPUTE GCD FROM THAT COMPUTE LEAST COMMON DIVISOR

## IDEA ::

Compute the LCD using GCD and from that for a range of numbers we can compute the LCD for that much range of numbers 

LCM of x,y,z is ==>  lcm(lcm(x,y),z)

```python
def gcd(x,y):
    while(y):
        x,y=y,x%y
    return x    
```

```python
gcd(300,400)
```

```python
gcd(24,54)
```

```python
def lcm(x,y):
    lcm=x*y//gcd(x,y)
    return lcm
```

```python
lcm(1,2)
```
